#### np.savetxt and np.loadtxt

```python
np.savetxt('savetxt.npy', arr, delimiter=',')
loaded_arr = np.loadtxt('savetxt.npy', delimiter=',')
loaded_arr
```

```python
with open('data.txt', 'w') as f:
    f.write('1, 2, 3\n')
    f.write('4,, 6\n')
    f.write('7, 8, 9\n')
```

```python
data = np.genfromtxt('data.txt', delimiter=',', filling_values=-1)
data
```

```python
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
```

```python
np.savez('my arrays.npz', arr1=arr1, arr2=arr2)
```

```python
loaded_data = np.load('my arrays.npz')
loaded_data['arr1']
```

```python
csv_data = '1, 2, 3\n4,,6\n7, 8, 9'
```

```python
data = np.genfromtxt(csv_data.splitlines(), delimiter=',', filling_values=-1)
data
```

When np.genfromtxt() processes this data, it will use the comma as the delimiter to separate values. So, for example, the first row will be parsed as [1. 2. 3.], the second row as [ 4. -1. 6.] (since the second value is missing, it is filled with the filling_values parameter, which is -1 in this case), and the third row as [7. 8. 9.].

In summary, the delimiter parameter specifies the character used to separate values in the input data when using np.genfromtxt().

1. `np.random.rand`: Generates random numbers from a uniform distribution between 0 and 1.
   Syntax: `np.random.rand(d0, d1, ..., dn)`
   Example: `np.random.rand(3, 4)` generates a 3x4 array of random numbers.

2. `np.random.randn`: Generates random numbers from a standard normal distribution (mean=0, std=1).
   Syntax: `np.random.randn(d0, d1, ..., dn)`
   Example: `np.random.randn(3, 4)` generates a 3x4 array of random numbers from a standard normal distribution.

3. `np.linspace`: Generates evenly spaced numbers over a specified interval.
   Syntax: `np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)`
   Example: `np.linspace(0, 10, num=5)` generates an array of 5 evenly spaced numbers between 0 and 10.

4. `np.arange`: Generates evenly spaced values within a given interval.
   Syntax: `np.arange([start, ]stop, [step, ]dtype=None)`
   Example: `np.arange(0, 10, 2)` generates an array of numbers from 0 to 10 with a step size of 2.
